# Selecionar subconjuntos do Dataframe

```python
coluna_tipo = pokemon_df["primary_type"] #selecionar uma coluna específica (tipo series, um array unidimensional)

type(coluna_tipo)
```

```python
coluna_tipo.tail()
```

```python
#e se eu quiser o tipo com o nome junto?
name_type_df = pokemon_df[["english_name", "primary_type", "secondary_type"]]

name_type_df.sample(5)
```

```python
type(name_type_df)
```

```python
#filtrar pokémon do tipo grama
grass_df = pokemon_df[pokemon_df["primary_type"] == 'grass']

grass_df.head()
```

```python
#filtrar pokémon fraco contra tipo água
weak_against_water_df = pokemon_df[pokemon_df["against_water"] > 2]

weak_against_water_df[["english_name", "primary_type", "against_water"]]
```

```python
#filtrar pokémon fraco contra tipo água e terra ao mesmo tempo
weak_against_water_ground_df = pokemon_df[(pokemon_df["against_water"] > 1) & (pokemon_df["against_ground"] > 1)]

weak_against_water_ground_df[["english_name", "primary_type", "against_water", "against_ground"]]
```

```python
#retornar pokémon com megaevolução
mega_df = pokemon_df[pokemon_df["mega_evolution"].notna()]

mega_df
```
